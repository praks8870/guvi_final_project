Reading and preprocessing the data

```python
df = pd.read_csv(r'influencer.csv')
```

```python
df
```

```python
def numbers(x):
    a =[]
    for f in x:
        if 'k' in f:
            n = round(1000*float(f.replace('k','')),0)
        if 'm' in f:
            n = round(1000000*float(f.replace('m','')),0)
        if 'b' in f:
            n = round(1000000000*float(f.replace('b','')),0)
        a.append(n)
    return a
```

```python
df['Followers_1'] = numbers(df['Followers'].to_list())
df['Avg. Likes_1'] = numbers(df['Avg. Likes'].to_list())
df['Posts_1'] = numbers(df['Posts'].to_list())
df['New Post Avg. Likes_1'] = numbers(df['New Post Avg. Likes'].to_list())
df['Total Likes_1'] = numbers(df['Total Likes'].to_list())

df['Channel Info'] = df['Channel Info'].str.replace('\n', '')
df
```

#   1. Are there any correlated features in the given dataset? If yes, state the correlation coefficient of the pair of features which are highly correlated.

In statistics, correlation measures the strength and direction of the relationship between two variables. A high correlation indicates a strong relationship between the variables, while a low correlation indicates a weak relationship. The correlation coefficient, denoted by r, ranges from -1 to 1, where Correlation coefficients whose magnitude are between 0.7 and 0.9 indicate variables which can be considered highly correlated.

```python
corr = df2.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
corr
```

```python
highly_correlated_pairs = (corr.abs() > 0.7) & (corr < 1)
highly_correlated_pairs
```

```python
#Features of pairs which are highly correlated

pair = [(feature1, feature2) for feature1 in column2 for feature2 in column2
                       if highly_correlated_pairs[feature1][feature2]]

x = []

for feature1, feature2 in pair:
    correlation_coefficient = corr.loc[feature1, feature2]
    x.append({'Feature1': feature1, 
               'Feature2': feature2, 
               'Correlation Coeff.': correlation_coefficient
               })

coeff_df = pd.DataFrame(x)
coeff_df
```

# 3. Which country houses the highest number of Instagram Influencers? Please show the count of Instagram influencers in different countries using barchart.

```python
plt.figure(figsize=(10, 8))
ax = sns.histplot(data=df1, x='Country Or Region', bins=  df1['Country Or Region'].nunique(), palette='viridis')

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='baseline', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')
    
plt.title('Count of Instagram Influencers based on Country')
plt.xlabel('Countries')
plt.ylabel('No. of Influencers')
plt.xticks(rotation=45, ha='right')
plt.show()
```
